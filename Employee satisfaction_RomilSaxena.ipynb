## Filtering All Employees with Attrition = No

```python
attrNoData = employeeData[(employeeData.Attrition == "No")]   
```

##  Further Filtering all "Attrition = No" employees with condition "BusinessTravel == "Travel_Frequently" as per problem statement

```python
AttrNoTravelFreq = attrNoData[(attrNoData.BusinessTravel == "Travel_Frequently")]  
```

```python
AttrNoTravelFreq
```

## Creating 3 Datasets for HR, Sales and R&D department with condition "Attrition = No" & "BusinessTravel == "Travel_Frequently"

```python
hrEmployees = AttrNoTravelFreq[(AttrNoTravelFreq.Department == "Human Resources")] 
SalesEmployees = AttrNoTravelFreq[(AttrNoTravelFreq.Department == "Sales")]
RnDEmployees = AttrNoTravelFreq[(AttrNoTravelFreq.Department == "Research & Development")]
```

# 2. 

## Below are the sales employees whose current JobRole is Sales Executive or Manager.

```python
SalesExecAndManagers = SalesEmployees[((SalesEmployees.JobRole == 'Sales Executive') | (SalesEmployees.JobRole == 'Manager'))]
SalesExecAndManagers
```

## Out of those who work as a Sales Executive or Manager, the company wants to give preference to those who are having PerformanceRating>=3, JobSatisfaction score >=3 and EducationField as "Marketing". On the basis of these conditions, report the shortlisted candidates who can be further interviewed by the company.

```python
SalesSelectedEmployees = SalesExecAndManagers[(SalesExecAndManagers.PerformanceRating>=3) & (SalesExecAndManagers.JobSatisfaction>=3) & (SalesExecAndManagers.EducationField == "Marketing")]
SalesSelectedEmployees
```

```python
print("Below Employee Numbers will be the shortlisted candidates as per criteria who can be further interviewed by the company "+'\n')
print ((SalesSelectedEmployees['EmployeeNumber']).to_string(index=False))
```

## Employees with above employee Number satisfies the criteria from Sales Department these employees can be shortlisted for further inteviews.
